```python
from skimage import io, color, morphology
from skimage.util import img_as_float, img_as_ubyte
import matplotlib.pyplot as plt
import numpy as np
import math
from skimage.filters import threshold_otsu
from skimage import segmentation
from skimage import measure
from skimage.color import label2rgb
```

```python
def show_comparison(original, modified, modified_name):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,
                                   sharey=True)
    ax1.imshow(original)
    ax1.set_title('Original')
    ax1.axis('off')
    ax2.imshow(modified)
    ax2.set_title(modified_name)
    ax2.axis('off')
    io.show()
```

```python
# From https://scikit-image.org/docs/stable/auto_examples/applications/plot_morphology.html
def plot_comparison(original, filtered, filter_name):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,
                                   sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(filter_name)
    ax2.axis('off')
    io.show()
```

```python
# Directory containing data and images
in_dir = "data/"

# X-ray image
im_name = "lego_4_small.png"

# Read the image.
# Here the directory and the image name is concatenated
# by "+" to give the full path to the image.
im_org = io.imread(in_dir + im_name)
```

```python
img_gray = color.rgb2gray(im_org)
```

```python
thresh = threshold_otsu(img_gray)
bin_img = img_gray < thresh
```

```python
plot_comparison(im_org, bin_img, 'Binary image')
```

```python
bin_img = segmentation.clear_border(bin_img)
```

```python
io.imshow(bin_img)
plt.title('Threshold image')
io.show()
```

```python
closed = morphology.closing(bin_img, morphology.disk(5))
opened = morphology.opening(closed, morphology.disk(5))
plot_comparison(bin_img, opened, 'closed & opened (disk5)')
```

```python
label_img = measure.label(opened)
n_labels = label_img.max()
print(f"Number of labels: {n_labels}")
```

```python
plot_comparison(bin_img, label2rgb(label_img), 'closed & opened (disk5)')
```

```python
region_props = measure.regionprops(label_img)
areas = np.array([prop.area for prop in region_props])
plt.hist(areas, bins=50)
plt.show()
```

## Cell counting

```python
in_dir = "data/"
img_org = io.imread(in_dir + 'Sample E2 - U2OS DAPI channel.tiff')
# slice to extract smaller image
img_small = img_org[700:1200, 900:1400]
img_gray = img_as_ubyte(img_small) 
io.imshow(img_gray, vmin=0, vmax=150)
plt.title('DAPI Stained U2OS cell nuclei')
io.show()
```

```python
# avoid bin with value 0 due to the very large number of background pixels
plt.hist(img_gray.ravel(), bins=256, range=(1, 100))
io.show()
```

## Exercise 8 - Threshold selection

```python
thres = 5
bin_img = img_as_ubyte(img_gray > thres)

plot_comparison(img_gray, bin_img, 'thres = 5')
```

```python
thres = threshold_otsu(img_gray)
bin_img = img_as_ubyte(img_gray > thres)

plot_comparison(img_gray, bin_img, "thres = 30 (otsu's)")
```

## Exercise 9 - Remove border BLOBS

```python
bin_img = segmentation.clear_border(bin_img)
plot_comparison(img_gray, bin_img, "thres = 30 (otsu's)")
```

```python
label_img = measure.label(bin_img)
image_label_overlay = label2rgb(label_img)
show_comparison(img_gray, image_label_overlay, 'Found BLOBS')
```

## Exercise 10 - BLOB features

```python
#compute a set of features for each object:
region_props = measure.regionprops(label_img)

#area of the first object be seen by:
print(region_props[0].area)
```

```python
#A quick way to gather all areas:
areas = np.array([prop.area for prop in region_props])
```

```python
print(areas)
```

```python
plt.hist(areas.ravel(), bins=len(areas), range=(60, 360))
io.show()
```

## Exercise 11 - BLOB classification by area

```python
min_area = 15
max_area = 150

# Create a copy of the label_img
label_img_filter = label_img
for region in region_props:
	# Find the areas that do not fit our criteria
	if region.area > max_area or region.area < min_area:
		# set the pixels in the invalid areas to background
		for cords in region.coords:
			label_img_filter[cords[0], cords[1]] = 0
# Create binary image from the filtered label image
i_area = label_img_filter > 0
plot_comparison(img_small, i_area, 'Found nuclei based on area')
```

## Exercise 12: Feature space

```python
#Extract all the perimeters of the BLOBS:
perimeters = np.array([prop.perimeter for prop in region_props])

#Try to plot the areas versus the perimeters:
plt.plot(areas, perimeters, '.b')
plt.xlabel("Area")
plt.ylabel('Perimeter')
plt.show()
```

## Exercise 13: BLOB Circularity

```python
circ = (4*np.pi*areas)/(perimeters**2)
plt.hist(circ, bins = len(circ))
plt.title("Circularity")
plt.show()
```

```python
print(circ)
```

## Exercise 14: BLOB circularity and area

```python
#Try to plot the areas versus the circularity:
plt.plot(areas, circ, '.b')
plt.xlabel("Area")
plt.ylabel('Perimeter')
plt.show()
```

```python
min_area = 15
max_area = 150
label_img_filter = label_img
num_objects = 0
for region in region_props:
    p = region.perimeter
    a = region.area
    circularity = (4*np.pi*a)/(np.square(p))
    if circularity < 0.8 or a < min_area or a > max_area:
        for cords in region.coords:
            label_img_filter[cords[0], cords[1]] = 0
    else:
        num_objects += 1

print("Objects: ", num_objects)
i_circ = label_img_filter > 0
show_comparison(img_small, i_circ, 'Found nuclei based on area and circularity')
```

```python
# slice to extract smaller image
img_slice = img_org[300:1200, 500:1400]
im_gray = img_as_ubyte(img_slice)
thres = threshold_otsu(im_gray)
img_bin = img_as_ubyte(im_gray > thres)
img_clear_border = segmentation.clear_border(img_bin)
label_img = measure.label(img_clear_border)
region_props = measure.regionprops(label_img)

min_area = 15
max_area = 150
label_img_filter = label_img
num_objects = 0
for region in region_props:
    p = region.perimeter
    a = region.area
    circularity = (4*np.pi*a)/(np.square(p))
    if circularity < 0.8 or a < min_area or a > max_area:
        for cords in region.coords:
            label_img_filter[cords[0], cords[1]] = 0
    else:
        num_objects += 1

print("Objects: ", num_objects)
i_circ = label_img_filter > 0
show_comparison(img_slice, i_circ, 'Found nuclei based on area and circularity')
```
